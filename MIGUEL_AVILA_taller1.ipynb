- ¿Qué?

Lo que se visualiza son areas distiguidas por colores de categorias de miedos/temores respecto a la intensidad  de menciones en los medios de comunicación entre los años 2000 y 2018. Se puede descubir, deducir,recordar y derivar las tendencias y comportamientos de los miedos a travez del tiempo y comparar con otros miedos. 

- ¿Por qué?

Porque el tipo de visulización empleado da una impresión completa de todos los datos a travez del tiempo implementado lo que permite identificar, localizar y comparar tendencias.

- ¿Cómo?

Codifica la información ordenada mediante marcas de area en un solo grafico mediante distinción de colores para cada categoria. 
Ademas al ser una visulización interactiva opcional permite que el usuario obtenga un resumen de cada temor si lo seleciona y tambien hacer dos cambios de escala del grafico y uno de perspectiva plana.

MilitarySpendingLN-superJumbo.png

- ¿Qué?

Se visualizan areas de circunferencia según la proporción tamaño/gasto del presupuesto militar por paises, se puede deducir  unicamente los 8 paises que mas gastan en su prespuesto militar tanto por los valores que lo marcan cada circulo y por los tamaños, colores de los circulos.Ademas el gasto en presupuesto de Estados Unidos es mayor tanto a los 7 paises que le siguen sumados y tambien es mayor al resto del gasto total de los demas paises.
parte negativa solo se puede conocer e identificar 8 paises de todos los circulos que hay ya que no tiene un identificador.  llegando a descubir, comparar, deducir y recordar información sustancial. 

- ¿Por qué?

la visulización da una impresión completa de solo los datos relevantes y permite identificar y comparar valores y tamaños de algunos paises

- ¿Cómo?

Codifica la información desordenada mediante marcas de area circular en un solo plano mediante distinción de una paleta de 3 colores para segmentar 3 conjuntos de paises por su presupuesto. no es visulización interactiva pero seria util para brindar información adicional. las barras inferiores son mas utiles ya que brindan mas facil y rapido una comprensión y comparación de los grupos.
